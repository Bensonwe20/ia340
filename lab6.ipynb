```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import matplotlib.pyplot as plt

# Filter the data for state 'VA'
df_va = df[df['name'] == 'Virginia']

# Calculate the income growth rate for VA
df_va = df_va.sort_values('year')
df_va['income_growth'] = df_va['income'].pct_change()

# Plot the income growth for VA
plt.figure(figsize=(10,6))
plt.plot(df_va['year'], df_va['income_growth'], marker='o', color='b')

plt.title('Income Growth Rate for Virginia')
plt.xlabel('Year')
plt.ylabel('Income Growth Rate')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataframe is named df
df['year'] = pd.to_numeric(df['year'])
df['income'] = pd.to_numeric(df['income'])

# Calculate the income growth rate
df['income_growth'] = df.groupby('name')['income'].pct_change()

# Pivot the dataframe to have years as columns, index as states and values as income growth
pivot_df = df.pivot(index='name', columns='year', values='income_growth')

# Plot the dataframe
plt.figure(figsize=(15,10))
for i in range(len(pivot_df)):
    plt.plot(pivot_df.columns, pivot_df.iloc[i], marker='o', label=pivot_df.index[i])
    
plt.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1)
plt.title('Income Growth Rate by State')
plt.xlabel('Year')
plt.ylabel('Income Growth Rate')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a pivot table with years as columns, states as rows, and income as values
pivot_df = df.pivot(index='name', columns='year', values='income')

# Calculate the growth rate (final value / initial value) - 1
growth_rate = (pivot_df[2019] / pivot_df[2005]) - 1

# Sort the states by growth rate
sorted_growth_rate = growth_rate.sort_values(ascending=False)

# Plot the growth rate for each state
plt.figure(figsize=(15, 10))
sorted_growth_rate.plot(kind='bar')
plt.title('Income Growth Rate by State from 2005 to 2019')
plt.ylabel('Growth Rate')
plt.xlabel('State')
plt.show()
```
